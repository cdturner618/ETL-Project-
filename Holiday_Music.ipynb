```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file="Resources/holiday_songs.csv"
df = pd.read_csv(file)
df.head()
```

```python
holiday_df=df[["track_name","artist_name","danceability"]].copy()
holiday_df
```

```python
holiday_df.columns = ("Song_Name", "Artist", "Danceability")
holiday_df.head()
```

```python
csv_file = "./Resources/featuresdf.csv"
music_df = pd.read_csv(csv_file)
music_df.head()
```

```python
new_music_df = music_df[['name', 'artists', 'danceability']].copy()
new_music_df.head()
```

```python
new_music_df.columns = ("Song Name", "Artist", "Danceability")
new_music_df.head()
```

```python
#Connection to local db

rds_connection_string = "postgres:postgres@localhost:5432/Holiday Music"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Check for tables
engine.table_names()
```

```python
holiday_df.to_sql(name='2017_Music', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from 2017_Music', con=engine).head()
```

```python
holiday_df.to_sql(name='Holiday_Music', con=engine, if_exists='append', index=False)
```
